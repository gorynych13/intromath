Рассчет длины вектора

```python
def vector_len(a):
    sum = 0
    for i in a:
        sum += i*i
    return np.sqrt(sum)

a = np.array([3, 4, 0])

vector_len(a)
```

```python
def circle_plot(a, b, r):
    """
    a, b : int, float
        zero point of circle 
    r : int, float
        radius
    """
    x = np.linspace(a - r, a + r, 101)
    y = np.sqrt(r*r -(x -a)**2) + b
    y1 = -np.sqrt(r*r -(x -a)**2) + b
    plt.plot(x, y, color='blue')
    plt.plot(x, y1, color='blue')
    plt.axis('equal')
    plt.show()
```

```python
circle_plot(101, 2, 50)
```

```python
def ellipce_plot(a, b, x0, y0):
    """
    x0, y0 : int, float
        zero point of circle 
    a, b : int, float
        
    """
    x = np.linspace(x0 - a, x0 + a, 101)
    y = np.sqrt(b**2 - b**2*(x-x0)**2/a**2) + y0
    y1 = -np.sqrt(b**2 - b**2*(x-x0)**2/a**2) + y0
    plt.plot(x, y, color='blue')
    plt.plot(x, y1, color='blue')
    plt.axis('equal')
    plt.show()
```

```python
ellipce_plot(5, 2, 0, 0)
```

```python
def giperbola_plot(a, b, x0, y0):
    """
    x0, y0 : int, float
        zero point of circle 
    a, b : int, float
        
    """
    x = np.linspace(-a*2, a*2, 301)
    y = np.sqrt(-b**2 + b**2*(x-x0)**2/a**2) + y0
    y1 = -np.sqrt(-b**2 + b**2*(x-x0)**2/a**2) + y0
    plt.plot(x, y, color='blue')
    plt.plot(x, y1, color='blue')
    plt.axis('equal')
    plt.show()
```

```python
giperbola_plot(3, 2, 0, 0)
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = X + 0.3*Y

ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z + 10)

show()
```
